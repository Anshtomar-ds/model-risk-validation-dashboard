# Credit Risk Model Monitoring

## Objective

The objective of model monitoring is to evaluate whether the model remains stable over time and to identify any data drift or performance degradation after deployment.

In this notebook, Population Stability Index (PSI) will be calculated to compare the training dataset with the validation dataset.

```python
# Import required libraries

import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")
```

```python
# Load training and validation datasets

train_df = pd.read_csv("../data/train.csv")
validation_df = pd.read_csv("../data/validation.csv")

print("Train Shape:", train_df.shape)
print("Validation Shape:", validation_df.shape)
```

## Population Stability Index (PSI)

Population Stability Index (PSI) is used to monitor whether the distribution of data has changed between the training dataset and the validation dataset.

PSI Interpretation:

- PSI < 0.10 → No significant drift
- 0.10 ≤ PSI < 0.25 → Moderate drift
- PSI ≥ 0.25 → Significant drift

```python
# Compare target distribution between train and validation datasets

train_target_dist = train_df["SeriousDlqin2yrs"].value_counts(normalize=True)
validation_target_dist = validation_df["SeriousDlqin2yrs"].value_counts(normalize=True)

psi_df = pd.DataFrame({
    "Train": train_target_dist,
    "Validation": validation_target_dist
})

psi_df
```

## PSI Calculation

The Population Stability Index (PSI) quantifies the change in distribution between the training dataset and the validation dataset.

A low PSI value indicates that the population remains stable and the model is likely to perform consistently.

```python
# Calculate PSI

psi = np.sum(
    (psi_df["Validation"] - psi_df["Train"]) *
    np.log(psi_df["Validation"] / psi_df["Train"])
)

print("PSI Value:", round(psi,6))
```

## PSI Interpretation

The calculated PSI value is 0.0000.

Since the PSI value is less than 0.10, there is no significant population drift between the training and validation datasets.

The model population remains stable and the model is expected to perform consistently.

```python
# Create PSI summary table

psi_summary = pd.DataFrame({
    "Metric": ["PSI"],
    "Value": [round(psi,6)],
    "Interpretation": ["No Significant Drift"]
})

psi_summary
```

```python
# Save monitoring report

psi_summary.to_csv(
    "../models/model_monitoring_report.csv",
    index=False
)

print("Monitoring Report Saved Successfully")
```
